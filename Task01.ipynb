```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"D:\Prodigy_Infotech\Data_01\train.csv")
df
```

```python
df[["LotArea","BsmtFullBath","BsmtHalfBath","FullBath","BedroomAbvGr","SalePrice"]]
```

```python
pd.isnull(df[["LotArea","BsmtFullBath","BsmtHalfBath","FullBath","BedroomAbvGr","SalePrice"]]).any()
```

Training dataset does not contain any blank cell. We can train our model and check the accuracy.

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(df[["LotArea","BsmtFullBath","BsmtHalfBath","FullBath","BedroomAbvGr"]],df[["SalePrice"]])
```

```python
model.score(df[["LotArea","BsmtFullBath","BsmtHalfBath","FullBath","BedroomAbvGr"]],df[["SalePrice"]])
```

Our training model has score around 41%.

```python
test = pd.read_csv(r"D:\Prodigy_Infotech\Data_01\test.csv")
test
```

```python
test_new= test[["LotArea","BsmtFullBath","BsmtHalfBath","FullBath","BedroomAbvGr"]]
test_new
```

```python
pd.isnull(test_new).any()
```

Here we can see some of the cells are blank. We need to delete those lines or put some value based on mean, median and mode.

```python
pd.isnull(test_new).sum()
```

only 4 cell have not any value inside it. let's resolve it be deleting those rows.

```python
mean_BsmtFullBath = test_new.BsmtFullBath.mean()
mean_BsmtHalfBath = test_new.BsmtHalfBath.mean()

test_new.BsmtFullBath = test_new.BsmtFullBath.fillna(mean_BsmtFullBath)
test_new.BsmtHalfBath = test_new.BsmtHalfBath.fillna(mean_BsmtHalfBath)
```

```python
pd.isnull(test_new).sum()
```

Now those rows which contain NaN value are dropped. So, Let predict the SalePrice of test data.

```python
Predicted_SalePrice = model.predict(test_new)
Predicted_SalePrice
```

```python
Predicted_SalePrice = pd.DataFrame(Predicted_SalePrice)
```

```python
Id = list(range(3000))
Id = Id[1461:2920]
print(np.shape(Id))
Id = pd.DataFrame(Id)
Id
```

```python
data = [Id,Predicted_SalePrice]
result = pd.concat(data, axis = 1)

column_names=["Id","SalePrice"]
result.columns = column_names

result
```
